주소를  로 변환하는 코드 
- http://api.vworld.kr 사이트의 (Geocoder API 2.0 레퍼런스) API사용


```python
import requests
import os

def request_geo(road,road_type,api_key):       #도로명주소, 지번주소, api키
    url = 'http://api.vworld.kr/req/address?'
    params = 'service=address&request=getcoord&version=2.0&crs=epsg:4326&refine=true&simple=false&format=json&type='
    address = '&address='
    keys = '&key='

    if road_type == "road": #도로명주소
        page = requests.get(url+params+road_type+address+road+keys+api_key)     # url + 파라미터 + 도로명주소 + api키
    elif road_type == "parcel": #지번주소
        page = requests.get(url+params+road_type+address+road+keys+api_key)     # url + 파라미터 + 지번주소 + api키   
    
    json_data = page.json()     #json 형태로 데이터 가져오기    
    if json_data['response']['status'] == 'OK':
        x = json_data['response']['result']['point']['x']     #x 
        y = json_data['response']['result']['point']['y']     #y 
        return x,y
    else:
        x = 0
        y = 0
        return 0, "주소를 찾을 수 없습니다. API-Key를 확인하세요."

#api키 가져오기
my_key = os.environ.get('MY_GEO_KEY')
if my_key is not None:
    print("환경변수 MY_GEO_KEY의 값을 가져왔습니다.")
else:
    print("환경변수 MY_GEO_KEY가 설정되어 있지 않습니다.")


x,y = request_geo(starbucks_df['소재지전체주소'].iloc[0],"parcel",my_key)
print(x,y)
```

```python
# 매장(스타벅스or이디야or메가), 매장이름, x, y  의 순으로 엑셀파일에 저장

starbucks_addr_list = starbucks_df['소재지전체주소'].to_list()
starbucks_name_list = starbucks_df['사업장명'].to_list()

ediya_addr_list = ediya_df['소재지전체주소'].to_list()
ediya_name_list = ediya_df['사업장명'].to_list()

mega_addr_list = mega_df['소재지전체주소'].to_list()
mega_name_list = mega_df['사업장명'].to_list()

shop_list = []
shop_name_list = []
x_list = []
y_list = []

#api키 가져오기
my_key = os.environ.get('MY_GEO_KEY')
if my_key is not None:
    print("환경변수 MY_GEO_KEY의 값을 가져왔습니다.")
else:
    print("환경변수 MY_GEO_KEY가 설정되어 있지 않습니다.")

for i,starbucks_addr in enumerate(starbucks_addr_list) :
    shop_list.append("스타벅스")
    shop_name_list.append(starbucks_name_list[i])
    try:
        x,y = request_geo(starbucks_addr,"parcel",my_key)
        x_list.append(x)
        y_list.append(y)
        print(starbucks_name_list[i],x,y)
    except:
        x_list.append(0)
        y_list.append(0)
        print("주소변환 에러")

for i,ediya_addr in enumerate(ediya_addr_list) :
    shop_list.append("이디야")
    shop_name_list.append(ediya_name_list[i])
    try:
        x,y = request_geo(ediya_addr,"parcel",my_key)
        x_list.append(x)
        y_list.append(y)
        print(ediya_name_list[i],x,y)
    except:
        x_list.append(0)
        y_list.append(0)
        print("주소변환 에러")

for i,mega_addr in enumerate(mega_addr_list) :
    shop_list.append("메가커피")
    shop_name_list.append(mega_name_list[i])
    try:
        x,y = request_geo(mega_addr,"parcel",my_key)
        x_list.append(x)
        y_list.append(y)
        print(mega_name_list[i],x,y)
    except:
        x_list.append(0)
        y_list.append(0)
        print("주소변환 에러")

all_df = pd.DataFrame()
all_df["매장"] = shop_list           # 매장 리스트 
all_df["매장이름"] = shop_name_list  # 매장이름 리스트
all_df["좌표_X"] = x_list
all_df["좌표_Y"] = y_list
all_df.to_excel("스타벅스_이디야_메가커피_위치.xlsx")
```
