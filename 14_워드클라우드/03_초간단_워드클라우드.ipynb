```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 텍스트
text = '''
휴넷은 7월 20일 오후 2시에 ‘ChatGPT가 불러온 HRD 혁신’을 주제로 라이브 세미나를 개최한다.

휴넷은 다달이 기업교육 관계자를 대상으로 하는 라이브 세미나 ‘월간 휴넷’을 진행하고 있다. 이번 행사는 기업교육의 핵심 화두를 선정해 최신 HRD 트렌드와 전략, 인사이트를 공유한다.

그동안 진행한 주제로는 ‘하이브리드 러닝’ ‘리더십 교육’ ‘승진자&신규 입사자 조기 전략화’ ‘러닝 저니’ ‘ATD 디브리핑’ 등이 있다. 기업교육 관계자들의 인기교육으로 각광 받으며 월평균 신청자 수는 1000여 명이 넘는다.

이번 세미나의 주제는 올해 화두인 ‘ChatGPT’이다.

사회적으로 엄청난 영향을 미치고 있는 주제인 만큼 기업의 관심도 뜨거울 것으로 전망된다. 강의에서는 ChatGPT와 같은 생성형 AI를 직무에 적용하고, 기업의 생산성을 높일 수 있는 방법을 제시할 예정이다.

휴넷 관계자는 “ChatGPT 관련 세미나가 많지만 대부분 이론에 그치는 경우가 많았다”며 “이번 세미나는 ChatGPT를 실무에 활용해 생산성을 높이는 방법, 새로운 비즈니스 기회 창출, 조직원들의 활용 필요성 등 기업 구성원들을 위한 효과적인 ChatGPT 활용과 교육전략 수립에 초점을 맞췄다”고 말했다.
'''

# 워드 클라우드 생성
wordcloud = WordCloud(font_path="NanumGothic", width=800, height=800, background_color='white', min_font_size=10).generate(text)

# 보여주기
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
import numpy as np
from PIL import Image # 이미지를 위한 라이브러리
import matplotlib.pyplot as plt # 그래프 및 이미지화 라이브러리

wine_mask = np.array(Image.open('img.jpg'))    # 이미지를 읽어와서 넘파이 배열로 저장
wordcloud = WordCloud(font_path="NanumGothic", background_color='white', mask=wine_mask, max_font_size=30).generate(text)

plt.figure()
plt.axis('off')
plt.imshow(wordcloud, interpolation='bilinear') 
plt.show()
```
