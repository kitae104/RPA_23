```python
from bs4 import BeautifulSoup
import requests
import os

h = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.212 Safari/537.36"}
response = requests.get("https://www.melon.com/chart/index.htm", headers=h)

html = response.text
soup = BeautifulSoup(html, 'html.parser')

singers = soup.select("span.checkEllipsis")
titles = soup.select(".ellipsis.rank01")
images = soup.select("a.image_typeAll > img")

# 이미지 저장 폴더 생성
image_folder = 'images'
if not os.path.exists(image_folder):
    os.makedirs(image_folder)

print(len(singers))
for i in range(len(singers)):
  singer = singers[i].get_text().strip()
  title = titles[i].get_text().strip()
  img_url = images[i]['src']

  # 이미지 파일 이름 설정
  image_filename = os.path.join(image_folder, f"{i + 1}_{singer}.jpg")

  image_response = requests.get(img_url, headers=h)
  with open(image_filename, 'wb') as f:
    f.write(image_response.content)

  print(f"{i+1}위 :  {title} / {singer} ")

```
