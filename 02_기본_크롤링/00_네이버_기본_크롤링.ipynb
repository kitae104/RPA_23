Beautifulsoup 사용 

```python
import requests
from bs4 import BeautifulSoup

response = requests.get("https://www.naver.com")
html = response.text
soup = BeautifulSoup(html, 'html.parser')
word = soup.select_one("#NM_set_home_btn")  # id는 #
print(word)
print(word.text)
```
