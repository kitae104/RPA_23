Beautifulsoup 사용 

```python
import urllib.request
from bs4 import BeautifulSoup
import ssl

context = ssl._create_unverified_context()
url = "https://www.naver.com"

html = urllib.request.urlopen(url,context=context).read()
soup = BeautifulSoup(html,'html.parser')

service_name = soup.select("span.blind")
print(service_name)

for i in service_name:
    print(i.text)
```

```python
import requests
from bs4 import BeautifulSoup

URL = "https://www.daum.net/"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

response = requests.get(URL, headers=headers)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')        
    # 2021년 9월 기준으로 네이버 메뉴 항목의 클래스 이름을 기반으로 가져옵니다.
    # 이후 웹사이트 구조 변경에 따라 클래스 이름 등이 변경될 수 있습니다.
    menu_items = soup.select('#daumHead > div.d_service > div > div.menu_service > ul > li:nth-child(1) > a > span')
    print(menu_items)
    for item in menu_items:
        print(item.get_text().strip())
else:
    print("Failed to retrieve the webpage.")
```
